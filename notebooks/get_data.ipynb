```python
import os
import pandas as pd
import glob
```

```python
data_dir = os.path.join("..", 'data')
dfile = os.path.join(data_dir, 'caltech_256_objects.tar')
os.path.isdir(data_dir)
```

```python
caltech_256_dset = 'https://www.vision.caltech.edu/Image_Datasets/Caltech256/256_ObjectCategories.tar'
```

```python
%%bash -s "$dfile" "$caltech_256_dset"
wget --no-check-certificate -O $1 $2
```

```python
%%bash -s "$dfile"
tar -xvf $1
```

```python
%%bash -s "$data_dir"
mkdir -p "$1/caltech_256"
mv "256_ObjectCategories" "$1/caltech_256/256_ObjectCategories"
```

```python
# make a reference file
all_image_files = glob.glob(os.path.abspath(os.path.join('..', 'data', 'caltech_256','256_ObjectCategories', '*', '*.jpg')))
```

```python
all_image_files
```

```python
im_ref_dataframe = pd.DataFrame(index=list(range(len(all_image_files))), columns=['class_name', 'class', 'file_loc'])
```

```python
len(all_image_files)
```

```python
common_root = os.path.commonprefix(all_image_files)
```

```python
common_root
```

```python
test = all_image_files[0]
test_end = test.replace(common_root, '')
test_end_dir, test_end_file = os.path.split(test_end)
print(test_end_dir, test_end_file)
test_class, test_class_name = test_end_dir.split('.')
print(test_class, test_class_name)
```

```python
for enum, im_file in enumerate(all_image_files):
    path_end = im_file.replace(common_root, '')
    path_end_dir, path_end_file = os.path.split(path_end)
    path_end_class, path_end_class_name = path_end_dir.split('.')
    path_end_class = int(path_end_class)
    
    im_ref_dataframe.set_value(enum, 'file_loc', im_file)
    im_ref_dataframe.set_value(enum, 'class', path_end_class)
    im_ref_dataframe.set_value(enum, 'class_name', path_end_class_name)
    
    
```

```python
outfile = os.path.join('..', 'data', 'cal256_file_reference.csv')
im_ref_dataframe.to_csv(outfile)
```
